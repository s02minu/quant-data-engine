```python
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf

data = yf.download(
    tickers= "BTC-USD",
    start='2023-01-01',
    interval='1d',
    group_by='ticker'

)
```

```python
data.head()
```

```python
df = pd.DataFrame(data)
print(f'The length of the data frame is {df.shape}')
df.head()
```

```python
print(df.columns.to_list())

```

```python
df.columns = ['Open', 'High', 'Low', 'Close', 'Volume']
print(df.columns.to_list())
```

```python
highest_close = df['Close'].max()
date_of_highest_close = df['Close'].idxmax()

print(highest_close)
print(date_of_highest_close)
```

```python

y = df['Close']


plt.figure(figsize = (14,5))
plt.plot(df.index, y)

plt.savefig("btc_close.png", dpi=300, bbox_inches='tight', transparent=True)


plt.xlabel("Date")
plt.ylabel("Price")
plt.title("BTC-USD Daily Close")
plt.show()

```
